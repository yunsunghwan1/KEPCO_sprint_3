```python
# 꺾은 선 그래프 그리기
locconelec_df.plot(figsize=(20, 10))
plt.legend(loc='upper right', bbox_to_anchor = (1, 0.9))
plt.title('연도별 행정구역별 전력 소비량')
plt.show()
```

```python
# 연도별 인구 수 추출
locconelec_year = []
for i in range(15) :
    locconelec_year.append(locconelec_df.iloc[i:i+1])
```

```python
locconelec_year[5]
```

```python
# 대한민국 지도 데이터 불러오기
jsonfile = open('sprint2_base/TL_SCCO_CTPRVN.json', 'r', 
                encoding='utf8').read()
jsondata = json.loads(jsonfile)
```

```python
### 지도맵 그리기
electronic_map_2012 = folium.Map(
    ### 최초에 보여줄 지도위치(위/경도) 지정
    # - 최초에 중심점을 기준으로 지도가 그려짐
    location = [37.573050, 126.979189],

    ### 지도 스타일 지정하기
    # - 도시형 건물, 산림, 하천/도로 등 스타일 지정
    # openstreetmap : 도시형 건물 스타일(가장 일반적으로 사용됨)
    tiles = "openstreetmap",

    ### 최초에 화면에 보여질 스케일(zoom) 지정하기
    zoom_start = 6.5
)

electronic_map_2012
```

```python
# 지역별 구분 시각화
folium.GeoJson(jsondata, name='시도').add_to(electronic_map_2012)
electronic_map_2012
```

```python
# 데이터 형태 회전
locconelec_2012_df = locconelec_year[5].T
```

```python
locconelec_2012_df
```

```python
# 데이터 컬럼 형태 확인
locconelec_2012_df.columns
```

```python
# 데이터 이름 재설정 및 수정
locconelec_2012_df.rename(columns={2012:'전력판매량'}, inplace=True)
```

```python
locconelec_2012_df
```

```python
locconelec_2012_df.index
```

```python
# 인덱스 초기화하여 인덱스로 되어있는 지역명을 칼럼으로 변환
locconelec_2012_df = locconelec_2012_df.reset_index()
```

```python
# 'index' 명을 '지역명'으로 컬럼명 재설정
locconelec_2012_df = locconelec_2012_df.rename(columns={'index': '지역명'})
```

```python
locconelec_2012_df
```

```python
# columns = ['서울시', '부산시', '대구시', '인천시', '광주시', '대전시', '울산시', '세종시', 
#            '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주도'], # DataFrame의 어떤 columns을 넣을지
```

```python
folium.Choropleth(
    geo_data = jsondata, # 경계선 좌표값이 담긴 데이터
    data = locconelec_2012_df, # Series or DataFrame 넣으면 된다
    columns = ['지역명','전력판매량' ], # DataFrame의 어떤 columns을 넣을지
    key_on = 'feature.properties.CTP_KOR_NM', # id 값을 가져오겠다; feature.id : feature 붙여줘야 함 (folium의 정해진 형식)
    fill_color = 'YlOrRd',
    fill_opacity = 0.6, # 색 투명도
    line_opacity = 0.6, # 선 투명도
    legend_name = '2012년 전력판매량', # 범례
    highlight=True,
).add_to(electronic_map_2012)
electronic_map_2012
```

```python
locconelec_2012_df
```

```python
# 전력판매량 기준으로 내림차순으로 정리
locconelec_2012_df = locconelec_2012_df.sort_values(by='전력판매량')

```

```python
locconelec_2012_df
```

```python
# 지역별 전력판매량 내림차순으로 바그래프로 변
# 데이터 준비
x = locconelec_2012_df['지역명']
y = locconelec_2012_df['전력판매량']

# 바 그래프 그리기
plt.barh(x, y)

# 그래프 제목 설정
plt.title('지역별 전력판매량')

# x축, y축 라벨 설정
plt.xlabel('전력판매량')
plt.ylabel('지역명')

# 그래프 크기 조정
plt.figure(figsize=(20, 20))

# 그래프 출력
plt.show()
```
